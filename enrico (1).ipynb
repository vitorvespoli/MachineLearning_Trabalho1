### Função para plotar os gráficos

```python
def plot_technical(
    x,
    y,
    title="Título do Gráfico",
    xlabel="Eixo X",
    ylabel="Eixo Y",
    labels=None,
    linewidth=2.0,
    markers=None,
    figsize=(8, 5),
    grid=True,
    save=False,
    filename="grafico.png",
    dpi=300
):
    plt.figure(figsize=figsize)

    # Detecta se y é curva única ou lista de curvas
    if isinstance(y, (pd.Series, np.ndarray, list, tuple)):
        if len(y) > 0 and not isinstance(y[0] if isinstance(y, (list, tuple, np.ndarray)) else y.iloc[0], (list, tuple, np.ndarray, pd.Series)):
            y = [y]
    else:
        y = [y]

    n_curves = len(y)

    if labels is None:
        labels = [f"Curva {i+1}" for i in range(n_curves)]

    if markers is None:
        markers = [None] * n_curves

    for i in range(n_curves):
        plt.plot(
            x,
            y[i],
            label=labels[i],
            linewidth=linewidth,
            marker=markers[i]
        )

    plt.title(title, fontsize=13)
    plt.xlabel(xlabel, fontsize=11)
    plt.ylabel(ylabel, fontsize=11)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    if grid:
        plt.grid(True, linestyle='--', alpha=0.7)

    if labels is not None:
        plt.legend(fontsize=10)

    plt.tight_layout()

    if save:
        plt.savefig(filename, dpi=dpi, bbox_inches='tight')

    plt.show()
```

## Análise Exploratória de Dados
Após a coleta e tratamento dos dados, esses foram segmentados em duas seções de transitórios (transitório inicial e final) e uma seção de regime estacionário, em que o manche do motor permanece no limite máximo, corroborando no máximo consumo de potência disponível conforme as capacidades do sistema.

A função abaixo realiza a identificação dessas seções. Primeiramente, há a aquisição da média referente aos valores entre 25% e 75% dos dados (seção aproximada de regime permanente), após isso, o algoritmo realiza uma varredura nas linhas dos dados de corrente em Ampères, sendo um dos principais parâmetros direcionados ao comportamento do sistema. A varredura ocorre visando encontrar os pontos de chaveamento de seções. O chaveamento entre o transitório inicial e o regime permanente é quando há o encontro de um dado entre 90 e 130% da média obtida. Já o chaveamento entre o regime permanente e o transitório final ocorre quando há o encontro de um dado abaixo de 82% do valor da média obtida. A busca pelo ponto de chaveamento entre o regime permanente e transitório final ocorre somente se o primeiro ponto de chaveamento já estiver obtido. Os valores de chaveamento foram obtidos a partir de testes.

```python
def transitory_stacionary(df):
    
    loc_i = df.columns.get_loc("Corrente_A") + 1

    q1 = df["Corrente_A"].quantile(0.30)
    q3 = df["Corrente_A"].quantile(0.70)

    df_50 = df[(df["Corrente_A"] >= q1) & (df["Corrente_A"] <= q3)]

    flag_stac = False
    transitory_2 = None
    flag_transitory_1 = True

    mean = df_50["Corrente_A"].mean()
    

    for idx, row in enumerate(df.itertuples()):
        if ((np.abs(row[loc_i]) >= 0.99*mean) & (np.abs(row[loc_i]) <= 1.3*mean) & flag_transitory_1):
            transitory_1 = idx + 10
            flag_stac = True
            flag_transitory_1 = False
            
        if flag_stac:
            if ((np.abs(row[loc_i]) <= 0.82*mean) & (np.abs(row[loc_i]) >= 0.3*mean)):
                transitory_2 = idx - 8
                break

    df_0 = df.iloc[:transitory_1]
    if transitory_2 != None:
        df_1 = df.iloc[transitory_1:transitory_2]
        df_2 = df.iloc[transitory_2:]
    else:
        df_1 = df.iloc[transitory_1:]
        df_2 = None

    return df_0, df_1, df_2

df_ex = pd.read_excel(os.path.join(quebradaFolder, "helice_quebrada_4.xlsx"))
df0, df1, df2 = transitory_stacionary(df_ex)

plot_technical(df_ex.index, df_ex["Corrente_A"], title="DataFrame completo", xlabel="Index", ylabel="Corrente [A]")
plot_technical(df0.index, df0["Corrente_A"], title="Transitório Inicial", xlabel="Index", ylabel="Corrente [A]")
plot_technical(df1.index, df1["Corrente_A"], title="Regime Permanente", xlabel="Index", ylabel="Corrente [A]")
plot_technical(df2.index, df2["Corrente_A"], title="Transitório Final", xlabel="Index", ylabel="Corrente [A]")
```

Os parâmetros considerados relevantes para a análise do sistema foram organizados por meio da classe statisticsParams, definida na célula abaixo. O objetivo é que cada objeto seja um tipo de operação, armazenando informações da média, mediana, valor máximo, valor mínimo, desvio padrão, skewness e kurtosis para cada coluna presente na operação. Além desses valores, é possível obter também a quantidade de dados para os transitórios e regime permanente.

```python
class statisticsParams:
    def __init__(self, name = None, df = None, tr1_qty = None, regime_qty = None, tr2_qty = None):
        
        mean = {}
        median = {}
        max = {}
        min = {}
        desv_padr = {}
        skew = {}
        kurtosis = {}
        mode = {}

        if df is not None:
            header = df.columns.to_list()    

            for h in header:
                mean[h] = df[h].mean()
                median[h] = df[h].median()
                max[h] = df[h].max()
                min[h] = df[h].min()
                desv_padr[h] = st.stdev(df[h])
                skew[h] = df[h].skew()
                kurtosis[h] = df[h].kurtosis()
                mode[h] = st.mode(df[h])

            self.name = name
            self.header = header
            self.mean = mean
            self.median = median
            self.max = max
            self.min = min
            self.desv_padr = desv_padr
            self.skew = skew
            self.kurtosis = kurtosis
            self.mode = mode

            self.tr1_qty = tr1_qty
            self.regime_qty = regime_qty
            self.tr2_qty = tr2_qty

        else:
            self.name = ""
            self.mean = mean
            self.median = median
            self.max = max
            self.min = min
            self.desv_padr = desv_padr
            self.skew = skew
            self.kurtosis = kurtosis
            self.mode = mode

            self.tr1_qty = 0
            self.regime_qty = 0
            self.tr2_qty = 0
```

A função abaixo é realizada com o objetivo de agrupar todas as coletas de dados realizadas em um tipo de operação, armazenando-as em um objeto "total", com as informações médias dos parâmetros estatísticos avaliados anteriormente.

```python
def statisticsGroup(ops, name):
    header = ops[0].header
    
    stat_temp = statisticsParams()
    

    for h in header:
        tr1_qty = 0
        tr2_qty = 0
        regime_qty = 0

        mean = []
        median = []
        maximum = []
        minimum = []
        desv_padrao = []
        skew = []
        kurtosis = []
        mode = []

        for op in ops:
            tr1_qty += op.tr1_qty
            tr2_qty += op.tr2_qty
            regime_qty += op.regime_qty

            mean.append(op.mean[h])
            median.append(op.median[h])
            maximum.append(op.max[h])
            minimum.append(op.min[h])
            desv_padrao.append(op.desv_padr[h])
            skew.append(op.skew[h])
            kurtosis.append(op.kurtosis[h])
            mode.append(op.mode[h])

        stat_temp.mean[h] = st.mean(mean)
        stat_temp.median[h] = st.mean(median)
        stat_temp.max[h] = max(maximum)
        stat_temp.min[h] = min(minimum)
        stat_temp.desv_padr[h] = st.mean(desv_padrao)
        stat_temp.skew[h] = st.mean(skew)
        stat_temp.kurtosis[h] = st.mean(kurtosis)
        stat_temp.mode[h] = st.mean(mode)
    
    stat_temp.name = name
    stat_temp.header = header
    stat_temp.tr1_qty = tr1_qty
    stat_temp.tr2_qty = tr2_qty
    stat_temp.regime_qty = regime_qty

    return stat_temp
```

Com o objetivo de visualizar as informações relevantes, a função abaixo foi definida, cujo argumento é um objeto da classe statisticParams, definida anteriormente. A função irá imprimir no terminal os valores armazenados no objeto.

```python
def stats_dataFrame(class_param):
    
    resumo = pd.DataFrame({
            "parameter": (h for h in class_param.header),
            "media": (class_param.mean[h] for h in class_param.header),
            "median": (class_param.median[h] for h in class_param.header),
            "mode": (class_param.mode[h] for h in class_param.header),
            "maximum": (class_param.max[h] for h in class_param.header),
            "minimum": (class_param.min[h] for h in class_param.header),
            "desvio_padrao": (class_param.desv_padr[h] for h in class_param.header),
            "skewness": (class_param.skew[h] for h in class_param.header),
            "kurtosis": (class_param.kurtosis[h] for h in class_param.header)
        })
    
    print("transitorio 1, quantidade de dados = ", class_param.tr1_qty)
    print("regime, quantidade de dados = ", class_param.regime_qty)
    print("transitorio 2, quantidade de dados = ", class_param.tr2_qty)
    print(resumo)
```

 - # <span style="color:yellow;">7. Análise de Correlação: Pearson e Spearman

   Para validar a integridade dos dados coletados pelo sistema de aquisição e entender a relação física entre as variáveis de operação do motor, foi aplicada uma análise estatística de correlação. O objetivo principal nesta etapa é quantificar a relação entre a potência (W) e a tração gerada pela hélice (Kgf), garantindo que o modelo de Machine Learning receba dados com coerência física.Para essa avaliação, foram utilizados dois métodos complementares: o Coeficiente de Pearson e o Coeficiente de Spearman.

- ### <span style="color:yellow;">7.1. Correlação de Pearson

   O coeficiente de Pearson foi aplicado para medir o grau de relação linear entre a potência e a tração. No geral , espera-se que o aumento da energia elétrica fornecida resulte em um aumento proporcional na tração. Valores próximos a 1 indicam que o sistema apresenta uma linearidade ideal, o que facilita a calibração de sensores e a previsibilidade do comportamento do conjunto motriz. Uma correlação de Pearson elevada confirma que o sistema de aquisição está operando com baixo ruído e percas.

- ### <span style="color:yellow;">7.2. Correlação de Spearman

  Diferente de Pearson, o coeficiente de Spearman avalia a relação monotônica entre as variáveis, baseando-se em ranks e não nos valores exatos, ou seja, ele ordena em ordem de maior para menor a tração e a potência e compara se o maior valor de potência é o maior valor de tração. A utilização deste método é estratégica para o projeto, pois ele é menos sensível a outliers e pequenas variações não-lineares. No contexto de falhas, como hélice quebrada ou desbalanceada, o fluxo aerodinâmico torna-se turbulento e a relação linear pode sofrer distorções. O coeficiente de Spearman permite verificar se as variáveis continuam crescendo juntas, mesmo que a eficiência do sistema tenha sido degradada pela falha mecânica.

Basicamente, qualquer subida na corrente força uma subida direta na tração. Então, mesmo que a hélice quebrada seja menos eficiente, a relação entre os dados ficou mais rígida e direta, o que acabou aumentando o valor do Spearman.

Nos dados da invertida, percebemos um valor pearson próximo com o da normal, garantindo a funcionalidade do motor tanto para rotação normal ou invertida, o dado de spearman foi maior devido a ausência de escoamento na captação de dado, visto que a rotação invertida muda o lado do escoamento.

 - ### <span style="color:yellow;">7.3. Análise complementar de Frequência via PSD (Densidade Espectral de Potência)

 A análise dos dados apenas no domínio do tempo e através de correlações lineares pode não ser suficiente para captar anomalias sutis na hélice. Para aprofundar o diagnóstico, aplicou-se a análise de Densidade Espectral de Potência (PSD) sobre os dados de tração, utilizando o Método de Welch.
 
Enquanto os coeficientes de Pearson e Spearman validam a relação entre a potência elétrica e a resposta mecânica (tração), a PSD nos permite visualizar como a energia vibratória está distribuída na frequência. Falhas estruturais como desbalanceamentos ou quebras alteram o "ritmo" de oscilação do conjunto GMP, criando picos de energia em frequências específicas que a análise estatística convencional ignora.



```python
######Ànalise_PSD##########

fs=10

sinal_tracao = df_20ordenado['Celula1_Kgf'].values

frequencias, psd = signal.welch(sinal_tracao, fs, nperseg=256)

plt.figure(figsize=(10, 4))
plt.semilogy(frequencias, psd, color='red')
plt.title('Densidade Espectral de Potência (PSD) - Tração(Desbalanceada)')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Densidade')
plt.grid(True, which='both')
plt.show()

sinal_tracao = df_10ordenado['Celula1_Kgf'].values

frequencias, psd = signal.welch(sinal_tracao, fs, nperseg=256)

plt.figure(figsize=(10, 4))
plt.semilogy(frequencias, psd, color='red')
plt.title('Densidade Espectral de Potência (PSD) - Tração(normal)')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Densidade')
plt.grid(True, which='both')
plt.show()

sinal_tracao = df_30ordenado['Celula1_Kgf'].values

frequencias, psd = signal.welch(sinal_tracao, fs, nperseg=256)

plt.figure(figsize=(10, 4))
plt.semilogy(frequencias, psd, color='red')
plt.title('Densidade Espectral de Potência (PSD) - Tração(quebrada)')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Densidade')
plt.grid(True, which='both')
plt.show()

sinal_tracao = df_40ordenado['Celula1_Kgf'].values

frequencias, psd = signal.welch(sinal_tracao, fs, nperseg=256)

plt.figure(figsize=(10, 4))
plt.semilogy(frequencias, psd, color='red')
plt.title('Densidade Espectral de Potência (PSD) - Tração(invertida)')
plt.xlabel('Frequência (Hz)')
plt.ylabel('Densidade')
plt.grid(True, which='both')
plt.show()
```

No gráfico da hélice Normal, observa-se que a maior concentração de energia está situada na frequência de 0 Hz. Isso é o comportamento ideal, pois indica que a força gerada pelo GMP é predominantemente contínua e estável. Ao longo de 1 a 5 Hz, o nível de ruído permanece baixo, demonstrando pouca vibração externa. 

Na hélice desbalanceada, nota-se uma mudança clara na distribuição de energia. Embora o pico em 0 Hz ainda exista, há um aumento na densidade de potência em frequências intermediárias. Visualmente, a curva torna-se mais serrilhada e com uma amplitude maior. Isso indica que o desequilíbrio de massa na hélice está convertendo parte da energia que deveria ser tração em oscilações mecânicas vibrátorias, o que justifica a queda nos coeficientes de correlação observados anteriormente.

Já na quebrada, observa-se o comportamente parecido com os anteriores, porem com a amplitude nas frequências de 1 a 5 Hz maiores que a normal, e menores que a desbalaceada, porém observamos um comportamente mais "agressivo" nas curvas, com uma variação grande em pequenos intervalos de frequência.

Na invertida, observamos um comportamento proximo ao da normal em referência a densidade, porem um pouco deslocado.
